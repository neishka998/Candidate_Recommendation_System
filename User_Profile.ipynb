```python
candidate_df = all_profile_df.copy()
#candidate_df = pd.read_csv("/content/sample_data/All_candidate_full_profiles.csv")
#candidate_df = candidate_df.loc[:, ~candidate_df.columns.str.contains('^Unnamed: 0')] #Removing column name "Unnamed"

candidate_df = candidate_df.drop(columns= ['Name', 'Current_Company', 'Job_Title', 'Company_2', 'Job_Title2', 'Job_Description_2','University', 'Degree_Name'])
print("Shape of the dataframe: ", candidate_df.shape)
print("\n")
print("." *500)
print("\n")
candidate_df.head()
```

```python
# Making all the job date range in the same naming conventions to calculate years of experience 
candidate_df.Job_Date_Range.iloc[58] = "Jun 2020 – Present"
candidate_df.Job_Date_Range2.iloc[58] = "Mar 2017 – May 2020"

candidate_df.Job_Date_Range.iloc[175] = "Jul 2020 – Present"
candidate_df.Job_Date_Range2.iloc[175] = "May 2016 – Jun 2020"

print("Changes are as follows : \n")
print(candidate_df.iloc[58])
print("." * 100)
print(candidate_df.iloc[175])
```

```python
# Calculating years of experience

# creating the date object of today's date
date = date.today()

for index, row in enumerate(candidate_df['Job_Date_Range']):
    if type(row) != float:
        num = int(re.findall(r'\d+', row)[0])
        candidate_df.loc[index, 'Job_Date_Range'] = date.year - num


for index, row in enumerate(candidate_df['Job_Date_Range2']):
    if type(row) != float:
        num_list = re.findall(r'\d+', row)
        if len(num_list) > 1:
            num = int(num_list[1]) - int(num_list[0])
        else:
            num = date.year - int(num_list[0])
        candidate_df.loc[index, 'Job_Date_Range2'] = num

candidate_df['Years_of_Exper'] = candidate_df['Job_Date_Range'] + candidate_df['Job_Date_Range2']
candidate_df = candidate_df.drop(columns=["Job_Date_Range", "Job_Date_Range2"])
# Rearranging the columnns 
candidate_df = candidate_df[['userId', 'Years_of_Exper', 'Degree', 'AllSkills']]
candidate_df
```

```python
# Checking for null values in the dataframe
print("Null Values: \n ", )
print(candidate_df.isnull().sum())
# Shape of the dataframe before dropping null values 
print("\n Orginal shape of the dataframe: ", candidate_df.shape)
print("." * 100)
print("\n")

# Dropping null values in allSkills columns 
candidate_df = candidate_df[candidate_df['AllSkills'].notna()]

# Checking for null values in the dataframe after dropping them 
print("\n Null Values: \n ", )
print(candidate_df.isnull().sum())
# Shape of thee dataframe after dropping null values 
print("\n Shape of the dataframe after dropping the null values: ", candidate_df.shape)
# Resetting the index
candidate_df = candidate_df.reset_index(drop=True)
print("." * 100)
print("\n")
candidate_df.head()
```

```python
# Making all values AllSkills columns lowercase
candidate_df['AllSkills'] = candidate_df['AllSkills'].str.lower()

# Converting Years of Experince to Seniority Level
def create_seniority_entry(dataframe, column):
  for index, exp in enumerate(dataframe[column]):
    if exp >= 5 :
      dataframe[column][index] = "Senior"
    elif exp <= 2:
      dataframe[column][index] = "Entry"
    else:
      dataframe[column][index] = "Mid"

create_seniority_entry(candidate_df, 'Years_of_Exper')
# Displaying the dataframe
candidate_df.head()
```
